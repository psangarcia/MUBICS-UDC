# Color

Por defecto en OpenCV las imágenes en color se cargan en formato BGR: el primer canal es el canal azul (B), el segundo canal es el verde (G) y el tercer canal, el rojo (R).

Para cambiar de espacio de color existe la función `cvtColor` que recibe como primer parámetro la imagen a transformar y, como segundo parámetro, una constante que representa el espacio de color actual de la imagen y el espacio de color de salida de la imagen. La salida de esta función es la imagen transformada al espacio de color elegido.

```python
%matplotlib inline
from matplotlib import pyplot as plt

import numpy as np
import cv2


im = cv2.imread('res/butterfly.jpg')

im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

plt.imshow(im_rgb)
plt.show()
```

Algunos de los posibles de cambios de espacio color son:

- GRAY: `cv2.COLOR_BGR2GRAY`, `cv2.COLOR_RGB2GRAY`, `COLOR_GRAY2BGR, COLOR_GRAY2RGB`
- XYZ: `cv2.COLOR_BGR2XYZ`, `cv2.COLOR_RBG2XYZ`, `cv2.COLOR_XYZ2RGB`, `cv2.COLOR_XYZ2BGR`, 
- YCrCb: `cv2.COLOR_BGR2YCrCb`, `cv2.COLOR_RBG2YCrCb`, `cv2.COLOR_YCrCb2BGR`, `cv2.COLOR_YCrCb2RGB`
- HSV: `cv2.COLOR_BGR2HSV`, `cv2.COLOR_RGB2HSV`, `cv2.COLOR_HSV2BGR`, `cv2.COLOR_HSV2RGB`
- HSL: `cv2.COLOR_BGR2HLS`, `cv2.COLOR_RGB2HLS`, `cv2.COLOR_HLS2BGR`, `cv2.COLOR_HLS2RGB`
- Lab: `cv2.COLOR_BGR2Lab`, `cv2.COLOR_RGB2Lab`, `cv2.COLOR_Lab2BGR`, `cv2.COLOR_Lab2RGB`
- Luv: `cv2.COLOR_BGR2Luv`, `cv2.COLOR_RGB2Luv`, `cv2.COLOR_Luv2BGR`, `cv2.COLOR_Luv2RGB`

## Ejercicios

1) Convierte la imagen de la mariposa de BGR a los espacios de color RGB, HSV y Lab. En cada uno de estos espacios de color:
a) Visualiza cada uno de los canales por separado
b) Comprueba cuáles son los valores mínimos y máximos de cada canal


```python
# Ejercicio 1a) Canales por separado para RGB

# Conversiones
im = cv2.imread('res/butterfly.jpg')

im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB) # Conversión a RGB
im_hsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV) # Conversión a HSV
im_Lab = cv2.cvtColor(im, cv2.COLOR_BGR2Lab) # Conversión a Lab

# Visualización de canales por separado en la imagen con espacio de color RGB
print('\n Ejercicio 1a) Visualiza cada uno de los canales por separado')
print('\n Visualización de canales por separado en la imagen con espacio de color RGB:')
plt.imshow(im_rgb[:,:,0], cmap='gray')
plt.show()
plt.imshow(im_rgb[:,:,1], cmap='gray')
plt.show()
plt.imshow(im_rgb[:,:,2], cmap='gray')
plt.show()

# Visualización de canales por separado en la imagen con espacio de color HSV
print('Visualización de canales por separado en la imagen con espacio de color HSV:')
plt.imshow(im_hsv[:,:,0], cmap='gray')
plt.show()
plt.imshow(im_hsv[:,:,1], cmap='gray')
plt.show()
plt.imshow(im_hsv[:,:,2], cmap='gray')
plt.show()

# Visualización de canales por separado en la imagen con espacio de color Lab
print('Visualización de canales por separado en la imagen con espacio de color Lab:')
plt.imshow(im_Lab[:,:,0], cmap='gray')
plt.show()
plt.imshow(im_Lab[:,:,1], cmap='gray')
plt.show()
plt.imshow(im_Lab[:,:,2], cmap='gray')
plt.show()
```

```python
# Ejercicio 1b) Comprueba cuáles son los valores mínimos y máximos de cada canal
print('\n Ejercicio 1b) Comprueba cuáles son los valores mínimos y máximos de cada canal \n')
print('Los valores máximos y mínimos para cada canal son: \n')
print('Imagen en RGB:')
print('canal R: ', np.min(im_rgb[:,:,0]), np.max(im_rgb[:,:,0]))
print('canal G: ', np.min(im_rgb[:,:,1]), np.max(im_rgb[:,:,1]))
print('canal B: ', np.min(im_rgb[:,:,2]), np.max(im_rgb[:,:,2]))
print('\n')
print('Imagen en HSV:')
print('canal H: ', np.min(im_hsv[:,:,0]), np.max(im_hsv[:,:,0]))
print('canal S: ', np.min(im_hsv[:,:,1]), np.max(im_hsv[:,:,1]))
print('canal V: ', np.min(im_hsv[:,:,2]), np.max(im_hsv[:,:,2]))
print('\n')
print('Imagen en Lab:')
print('canal L: ', np.min(im_Lab[:,:,0]), np.max(im_Lab[:,:,0]))
print('canal a: ', np.min(im_Lab[:,:,1]), np.max(im_Lab[:,:,1]))
print('canal b: ', np.min(im_Lab[:,:,2]), np.max(im_Lab[:,:,2]))
```

3) Determina cuál es el espacio de color (RGB, Lab o HSV) y el canal o canales que aportan más información en las siguientes imágenes: 

```python
# Análisis para la imagen retino.jpg:

im1 = cv2.imread('res/retino.jpg') # Detectar vasos

im1_rgb = cv2.cvtColor(im1, cv2.COLOR_BGR2RGB) # Se aplica conversión a RGB
im1_hsv = cv2.cvtColor(im1, cv2.COLOR_BGR2HSV) # Se aplica conversión a HSV
im1_Lab = cv2.cvtColor(im1, cv2.COLOR_BGR2Lab) # Se aplica conversión a Lab

# Muestra de espacios y canales correspondientes a la imagen retino.jpg:
plt.rcParams["figure.figsize"] = [16,16]
f, ax = plt.subplots(4,3)
ax[0,0].imshow(im1_rgb)
ax[0,0].set_title('RGB')
ax[0,1].imshow(im1_hsv)
ax[0,1].set_title('HSV')
ax[0,2].imshow(im1_Lab)
ax[0,2].set_title('Lab')
ax[1,0].imshow(im1_rgb[:,:,0], cmap='gray')
ax[1,0].set_title('R')
ax[1,1].imshow(im1_rgb[:,:,1], cmap='gray')
ax[1,1].set_title('G')
ax[1,2].imshow(im1_rgb[:,:,2], cmap='gray')
ax[1,2].set_title('B')
ax[2,0].imshow(im1_hsv[:,:,0], cmap='gray')
ax[2,0].set_title('H')
ax[2,1].imshow(im1_hsv[:,:,1], cmap='gray')
ax[2,1].set_title('S')
ax[2,2].imshow(im1_hsv[:,:,2], cmap='gray')
ax[2,2].set_title('V')
ax[3,0].imshow(im1_Lab[:,:,0], cmap='gray')
ax[3,0].set_title('L')
ax[3,1].imshow(im1_Lab[:,:,1], cmap='gray')
ax[3,1].set_title('a')
ax[3,2].imshow(im1_Lab[:,:,2], cmap='gray')
ax[3,2].set_title('b')
plt.show()


```

En base a los resultados obtenidos para cada espacio de color y canales en la imagen retino.jpg, se puede concluir que el espacio RGB y sus respectivos canales, son los que proporcionan una mayor información en la imagen.

```python
# Análisis para la imagen fish.jpg:
im2 = cv2.imread('res/fish.jpg') # Detectar células y marcadores

im2_rgb = cv2.cvtColor(im2, cv2.COLOR_BGR2RGB) # Se aplica conversión a RGB
im2_hsv = cv2.cvtColor(im2, cv2.COLOR_BGR2HSV) # Se aplica conversión a HSV
im2_Lab = cv2.cvtColor(im2, cv2.COLOR_BGR2Lab) # Se aplica conversión a Lab

# Muestra de espacios y canales correspondientes a la imagen fish.jpg:
plt.rcParams["figure.figsize"] = [16,16]
f, ax = plt.subplots(4,3)
ax[0,0].imshow(im2_rgb)
ax[0,0].set_title('RGB')
ax[0,1].imshow(im2_hsv)
ax[0,1].set_title('HSV')
ax[0,2].imshow(im2_Lab)
ax[0,2].set_title('Lab')
ax[1,0].imshow(im2_rgb[:,:,0], cmap='gray')
ax[1,0].set_title('R')
ax[1,1].imshow(im2_rgb[:,:,1], cmap='gray')
ax[1,1].set_title('G')
ax[1,2].imshow(im2_rgb[:,:,2], cmap='gray')
ax[1,2].set_title('B')
ax[2,0].imshow(im2_hsv[:,:,0], cmap='gray')
ax[2,0].set_title('H')
ax[2,1].imshow(im2_hsv[:,:,1], cmap='gray')
ax[2,1].set_title('S')
ax[2,2].imshow(im2_hsv[:,:,2], cmap='gray')
ax[2,2].set_title('V')
ax[3,0].imshow(im2_Lab[:,:,0], cmap='gray')
ax[3,0].set_title('L')
ax[3,1].imshow(im2_Lab[:,:,1], cmap='gray')
ax[3,1].set_title('a')
ax[3,2].imshow(im2_Lab[:,:,2], cmap='gray')
ax[3,2].set_title('b')
plt.show()
```

En comparación con el ejercicio anterior, para la imagen fish.jpg, los resultados parecen indicar que los espacios de color RGB y Lab son los que aportan más información. Por su parte, los canales siguen una tendencia similar, donde se aprecia que los canales R y G junto con el canal a, proporcionan una mayor información en la imagen.
